# Seismic Waves

## 1. Types of seismic waves

There are three main types of waves that generated by earthquakes and that are used to investigate the structure of a planet.

The figure that the code below makes shows a seismogram for an earthquake that happened in Alaska recorded at a seismometer in Scotland. It is showing how the waves from the earthquake cause the ground to move in the vertical (up and down) direction. 

```python
from obspy import UTCDateTime, read, Stream
from obspy.clients.fdsn import Client

#Information about the time of the earthquake
t_eq=UTCDateTime("2020-07-22T06:12:44.00")
c = Client('IRIS') # Client for data from IRIS

network = "GB" #network of seismometer
station = "DRUM" #station name
location = "*" #location 
channel = "BHZ" #all 3 broadband components 

UTCstartcut=t_eq # download from the time of the earthquake
UTCendcut=t_eq+4000 # download to 4000s after the earthquake 

st = c.get_waveforms(network, station, location, channel, UTCstartcut, UTCendcut, attach_response=True)
fig=st.plot()
```

The first waves to arrive, at xxxx, about XXXX seconds after the earthquake, are the P wave. The second waves to arrive, at xxxxx, about xxxxx seconds after the earthquake, are the S waves. These are known as body waves as they travel through a planet from the earthquake to the seismometer. The waves that arrive between xxxxx and xxxxxx are surface waves. These are formed from the interaction of P and S waves and travel along the surface of the planet.
